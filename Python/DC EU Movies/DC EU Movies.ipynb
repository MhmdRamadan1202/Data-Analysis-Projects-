# DCEU Box Office and Rating Dataset
##### DC Extended Universe Box Office and Rating Data

## About Dataset
#### Context


When I looked for an DCEU datasets on Kaggle, I coudl not found a proper dataset for data analysis and visualization. Since a fan of DC Extended Universe (DCEU), I thought it would be a fun to create an DCEU box office and rating dataset. This dataset includes box office datas of all DCEU movies, release date, movie duration, movie budget, US Gross, Worldwide Gross, imdb score, metascore, tomatometer score, and tomatometer audience score.
* movie_title -> Title of the movie
* release_date -> Release date of the movie
* duration (mins) -> Duration of the movie
* budget -> Production budget of the movie
* US_gross -> US Gross of the movie
* worldwide_gross -> Worldwide Gross of the movie
* imdb -> iMDb Score of the movie
* metascore -> Metascore score of the movie
* tomatometer -> Tomatometer score of the movie
* audience_score -> Tomatometer audience score of the movie

#### Data Extracted From:

* https://en.wikipedia.org/wiki/DC_Extended_Universe
* https://www.imdb.com/list/ls062501894/
* https://www.metacritic.com/company/dc-entertainment
* https://www.rottentomatoes.com/franchise/dc_comics
* https://www.boxofficemojo.com/franchise/fr340233989/

## Visualization Analysis

```python
sns.lineplot(movies)
```

```python
sns.lineplot(data=movies, x="movie_title", y="worldwide_gross ($)")
plt.xticks(rotation=90)
```

```python
df = movies.sort_values(by=['imdb'])
plt.plot(df["imdb"], df["worldwide_gross ($)"])
plt.xticks(rotation=90);
```

```python
movies.plot(x="movie_title", y=["budget ($)", "worldwide_gross ($)"], kind="bar")
plt.xticks(rotation=90);
```

```python
movies.plot(x="movie_title", y=["imdb", "metascore", "tomatometer", "audience_score"], kind="barh", figsize=(8,12))
plt.xticks(rotation=90);
```

```python
movies.plot.pie(y = "budget ($)")
```

```python
fig = plt.figure(figsize=(40,10), dpi=1600)
fig, (ax1, ax2) = plt.subplots(1, 2,  figsize=(20,20))
fig.tight_layout() 

ax1.pie(movies["budget ($)"], labels = movies["movie_title"]);
ax2.pie(movies["worldwide_gross ($)"], labels = movies["movie_title"]);
```

```python
plt.pie(movies["worldwide_gross ($)"], labels = movies["movie_title"]);
plt.show() 
```
